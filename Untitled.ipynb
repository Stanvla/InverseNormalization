```python
import sys


from Bio import pairwise2
from Levenshtein import distance
import json
from string import punctuation
import re
from collections import Counter


with open('data.json', 'r') as f:
    raw_data = json.load(f)
```

```python
ignore = [
    '“', # ignore 
    '„', # ignore
    '"', # ignore
    '(', # ignore
    ')', # ignore
    ' %', # ignore    
    '%', # ignore
    ' –', # ignore
    ' —',
    ' -',
    '-',
    ' &',
    '&', # ignore
    '”', # ignore
    '°', # ignore
]

data = dict()
for i in range(len(raw_data)):
    new_raw = raw_data[f'{i}']['raw']
    norm = raw_data[f'{i}']['norm']
    for s in ignore:
        new_raw = new_raw.replace(s, '')
    data[i] = dict(norm=norm, new_raw=new_raw)
```

```python
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def has_punc(my_text):
    return any(p in my_text for p in punctuation)

def get_punc(x):
    for p in punctuation:
        if p in x:
            return p

def match_fn(a, b):
    a = a.lower()
    b = b.lower()
    if any(char.isdigit() for char in b):
        return 0.1
    return max(len(a), len(b)) - distance(a, b)

def display_alignment(a, b, gap_char):
    result = []
    last = ''
    for i, (x, y) in enumerate(zip(a[::-1], b[::-1])):
        if x.lower() == y.lower():
            # print(f'{x:^55} | OK')
            if x != y:
                result.append('case')
            else:
                result.append('ok')
        else:
            r = f'{x:^25} ... {y:^25}'
            if has_numbers(y):
                if '.' in y and i != len(b) -1:
                    last = 'num_ord'
                    r = f'{r} | {last}'
                    result.append(last)
                else:
                    last = 'num_card'
                    r = f'{r} | {last}'
                    result.append(last)
            else:
                if has_punc(y) and y != gap_char:
                    if y.lower() != y:
                        last = f'case_punc_{get_punc(y)}'
                        r = f'{r} | {last}'
                        result.append(last)
                    else:
                        last = f'punc_{get_punc(y)}'
                        r = f'{r} | {last}'
                        result.append(last)
                elif y == gap_char:
                    if 'num' in last:
                        r = f'{r} | {last}'
                        result.append(f'{last}')
                    else:
                        result.append('ok')
                else:
                    r = f'{r} | ok'
                    result.append(f'ok')
            # print(r)
    return result[::-1]


data_aligned = dict()
errors = dict()

for i, v in data.items():
#     if has_numbers(v['raw']) and isinstance(v['norm'], str):
    if isinstance(v['norm'], list):
        continue
    alignment = pairwise2.align.globalxx(
        v['norm'].split(), 
        v['new_raw'].split(),
#             gap_char='@'
        gap_char=['~'],
        match_fn=match_fn,

    )[0]
    norm_align, raw_align, _, _, _ = alignment
    norm_align_str = ' '.join(norm_align)
    
    res = display_alignment(norm_align, raw_align, '~')
    if '~' not in norm_align_str:
        data_aligned[i] = dict(norm=norm_align, raw=raw_align, trg=res)
    else:
        errors[i] = dict(norm=norm_align, raw=raw_align, trg=res)
    
#     if '~' not in norm_align_str and len(norm_align) != len(v['norm'].split()): 
#         print(cnt)
#         print(f'norm :: {v["norm"]}')
#         print(f'raw  :: {v["raw"]}')
# #         print(norm_align)
# #         print(raw_align)
#         display_alignment(norm_align, raw_align, '~')
#         print('----'*20)
#         cnt += 1

    
```

```python
def update_counter(indices):
    lst = [data_aligned[i] for i in indices]
    trgs = [trg for d in lst for trg in d['trg']]
    return Counter(trgs)

import math
val_indices = []
test_indices = []
train_indices = []

val_counter = {k:0 for k in counter}
test_counter = {k: 0 for k in counter}

val_frac = 0.15
test_frac = 0.15
used_indices = []

for tag, cnt in counter.most_common()[::-1]:
    test_cnt = math.floor(test_frac * cnt)
    val_cnt = math.floor(val_frac * cnt)
    print(tag, cnt, val_cnt, test_cnt)
    
    for i, d in data_aligned.items():
        norm, trg = d['norm'], d['trg']
        if tag in trg and i not in used_indices:
            if val_cnt > val_counter[tag]:
                val_indices.append(i)
                val_counter[tag] += trg.count(tag)
                used_indices.append(i)
            elif test_cnt > test_counter[tag]:
                test_indices.append(i)
                test_counter[tag] += trg.count(tag)
                used_indices.append(i)
            else:
                break
    val_counter = update_counter(val_indices)
    test_counter = update_counter(test_indices)
                
```
